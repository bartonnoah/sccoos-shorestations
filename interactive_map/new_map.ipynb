```python
import folium
from folium import IFrame
import pandas as pd
from datetime import datetime
import branca.colormap as cm
from branca.element import Template, MacroElement
```

```python
#save constants
stations = pd.DataFrame()
stations['nice_name'] = ('Scripps Pier, La Jolla', 'San Clemente', 'Newport Beach', 'Point Dume', 'Santa Barbara', 
                    'Granite Canyon', 'Pacific Grove', 'Farallon Islands', 'Trinidad Beach & Bay')
stations['name'] = ('scripps', 'clemente', 'newport', 'zuma', 'barbara', 
                    'granite', 'grove', 'farallons', 'trinidad')
stations['lat'] = (32.8669444, 33.4190556, 33.606694, 34.017917, 34.403861, 36.439083, 36.6202482, 37.6964444, 41.0600833)
stations['lon'] = (-117.2597192, -117.6236081, -117.9327747, -118.827278, -119.692778, -121.922528, -121.9063325, -123.0021268, -124.1515803)
n_stations = len(stations)
sw = (32, -124)
ne = (42, -114)
```

```python
#create map and set extent
m = folium.Map(location=[35, -120], tiles='cartodbpositron')
m.fit_bounds([sw, ne])

#add station markers
for i in range(n_stations):
    name = stations.nice_name[i]
    lat = stations.lat[i]
    lon = stations.lon[i]
    with open('station_htmls/'+stations.name[i]+'.html') as f:
        popup_html = f.read()

    marker = folium.CircleMarker(location = [lat,lon],
                                 radius = 5,
                                 fill = True,
                                 fill_opacity = 1)
    
    tooltip = folium.Tooltip(name)
    marker.add_child(tooltip)
    popup = folium.Popup(IFrame(html=popup_html, width=450, height=400))
    marker.add_child(popup)
    marker.add_to(m)
m
```

```python
m.save('new_map.html')
```
